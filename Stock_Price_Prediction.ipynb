## 1. Problem Statement

The objective of this project is to predict the **next day's closing stock price** using historical market data obtained from Yahoo Finance. 
Machine learning regression techniques are applied to understand relationships between stock indicators such as Open, High, Low, and Volume.

This notebook demonstrates the complete workflow including:
- Data collection using an API
- Data preprocessing
- Exploratory data analysis
- Model training
- Evaluation
- Visualization of predictions

---

## 2. Project Goal

To build a regression model capable of estimating the next trading day's closing price based on historical stock performance.



## 3. Libraries Used

- pandas → data manipulation
- numpy → numerical operations
- matplotlib → visualization
- yfinance → stock data API
- scikit-learn → machine learning models


```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```


## 4. Data Collection

We retrieve historical stock market data using the **yfinance API**.
For this project, Tesla (TSLA) stock is selected as an example.


```

ticker = "TSLA"

data = yf.download(ticker, start="2018-01-01", end="2024-12-31")

data.head()
```


## 5. Data Exploration

Understanding dataset structure helps ensure correct preprocessing before model training.


```

data.info()
data.describe()
```


## 6. Visualization of Closing Prices

Plotting the closing price helps identify trends and volatility patterns over time.


```

plt.figure(figsize=(12,6))
plt.plot(data['Close'])
plt.title("Tesla Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```


## 7. Feature Engineering

To predict tomorrow's price, we shift the closing price column upward by one day.


```

data['Next_Close'] = data['Close'].shift(-1)
data = data.dropna()
```


## 8. Feature Selection

Input Features:
- Open
- High
- Low
- Volume

Target:
- Next day's Close price


```

features = ['Open', 'High', 'Low', 'Volume']
X = data[features]
y = data['Next_Close']
```


## 9. Train-Test Split

Since stock data is time-series, shuffling is disabled to preserve chronological order.


```

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)
```


## 10. Model Training — Linear Regression


```

model = LinearRegression()
model.fit(X_train, y_train)
```


## 11. Predictions


```

predictions = model.predict(X_test)
```


## 12. Model Evaluation

We evaluate performance using:
- Mean Squared Error (MSE)
- R² Score


```

mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print("Mean Squared Error:", mse)
print("R2 Score:", r2)
```


## 13. Actual vs Predicted Prices Visualization


```

plt.figure(figsize=(12,6))
plt.plot(y_test.values, label="Actual Price")
plt.plot(predictions, label="Predicted Price")
plt.title("Actual vs Predicted Closing Prices")
plt.xlabel("Days")
plt.ylabel("Stock Price")
plt.legend()
plt.show()
```


## 14. Predict Next Day Closing Price


```

latest_data = X.iloc[-1].values.reshape(1, -1)
next_day_prediction = model.predict(latest_data)

print("Predicted Next Day Closing Price:", next_day_prediction[0])
```


## 15. Results & Discussion (Student Reflection)

### Observations
- The model captures general market trends reasonably well.
- Predictions follow overall movement patterns but struggle with sudden spikes.

### Challenges
- Stock prices are influenced by external news and economic factors.
- Pure historical data limits prediction accuracy.

### Improvements for Future Work
